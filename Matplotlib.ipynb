# Area Plot

```python
days = [1,2,3,4,5]
sleeping = [6,4,5,7,8]
eating = [0.5,1,1.5,0.6,2.3]
working = [11,13,15,16,9]
playing = [1,1.2,3,2.4,2]

plt.plot([], [], color='m', label='sleeping', linewidth=5)
plt.plot([], [], color='g', label='eating', linewidth=5)
plt.plot([], [], color='r', label='working', linewidth=5)
plt.plot([], [], color='b', label='playing', linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing, colors= ['m','g','r','b'])
plt.legend()
```

# Pie Chart

```python
days = [1,2,3,4,5]
sleeping = [6,4,5,7,8]
eating = [0.5,1,1.5,0.6,2.3]
working = [11,13,15,16,9]
playing = [1,1.2,3,2.4,2]
slices=[7,3,2,18]

activities = ['sleeping','eating','working','playing']
cols = ['m','r','g','b']
plt.pie(slices, labels=activities,colors=cols,startangle=90,shadow=True)
```

# Drawing graphs on external data frame

```python
football = pd.read_csv('fifa_data.csv')
```

```python
football.head()
```

```python
football.info()
```

```python
bins = [20,22,24,26,28,30,32,34,36,38,40]
plt.figure(figsize=(8,5))
plt.hist(football.Age, bins=bins,color='#1c2042')
```

```python
left = football.loc[football['Preferred Foot'] == 'Left'].count()[0]

right = football.loc[football['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize = (8,5))

labels = ['Left', 'Right']

color = ['#1c2042', '#1a61f6']

plt.pie([left, right], labels = labels, colors = color)
```

```python
##Create a Pie chart for the Football data on weight of the Football Player

##Weight less than 125
##Weight between 125 and 150
## Weight between 150 and 175
## Weight between 175 and 200
## Weight above 200

## Properly Labeleed,  with Title
```

```python
w1 = football.loc[football['Weight']<'125lbs'].count()[0]
w2 = football.loc[football['Weight'].between('125lbs','150lbs')].count()[0]
w3 = football.loc[football['Weight'].between('150lbs','175lbs')].count()[0]
w4 = football.loc[football['Weight'].between('172lbs','200lbs')].count()[0]
w5 = football.loc[football['Weight']>'200lbs'].count()[0]

plt.figure(figsize = (8,5)) 

labels = ['w1','w2','w3','w4','w5']
                  
color = ['y', 'k','r','g','b']
                  
plt.pie([w1,w2,w3,w4,w5], labels = labels, colors = color)
```
