```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
#importing main dataset
df = pd.read_csv('covid_19_india.csv',parse_dates=['Date'], dayfirst=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#keeping only required columns
df = df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]

#renaming column names
df.columns = ['date','state','cured','deaths','confirmed']
```

```python
#looking at the earliest dates
df.head()
```

```python
#looking at the latest dates
df.tail()
```

```python
#current date
today = df[df.date=='2021-04-28']
today
```

```python
#sorting data w.r.t to no.of confirmed cases
max_confirmed_cases=today.sort_values(by='confirmed',ascending=False)
max_confirmed_cases
```

# India

```python
#Linear Regression-For Entire Country
from sklearn.model_selection import train_test_split
```

```python
#converting datetime to ordinal
df['date']=df['date'].map(dt.datetime.toordinal)
df.head()
```

```python
#Getting dependent and independent variable
x=df['date']
y=df['confirmed']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
df.tail()
```

```python
lr.predict(np.array([[737908]]))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dr=DecisionTreeRegressor()
```

```python
dr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
df.tail()
```

```python
dr.predict(np.array([[737908]]))
```

```python
#Checking accuracy
lr.score(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```
