# Lambda function inside a user defined function

```python
#a*x**2 + b*x + c
def evaluate(a,b,c):
    return lambda x:a*x**2+ b*x +c
result=evaluate(5,6,7)
print(result(4))
    
```

# Lambda function inside a default function

```python
ids=['id3','id11','id34','id25','id55']
sorted_ids=sorted(ids, key=lambda x: int(x[2:]))
print(sorted_ids)
```

```python
death=[('James',43),('Ramesh',24),('Kishore',42)]
sorted(death, key=lambda age:age[1])
```
