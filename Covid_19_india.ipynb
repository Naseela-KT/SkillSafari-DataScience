```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
#keeping only required columns
df = df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]

#renaming column names
df.columns = ['date','state','cured','deaths','confirmed']
```

```python
#current date
today = df[df.date=='2021-04-28']
today
```

```python
#sorting data w.r.t to no.of confirmed cases
max_confirmed_cases=today.sort_values(by='confirmed',ascending=False)
max_confirmed_cases
```

```python
#Getting states with maximum no.of confirmed cases
top_state_confirmed=max_confirmed_cases[0:5]
```

```python
#making bar-plot for states with top confirmed cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='state',y='confirmed',data=top_state_confirmed,hue='state')
plt.show()
```

```python
#Linear Regression
from sklearn.model_selection import train_test_split
```

```python
#converting datetime to ordinal
maha['date']=maha['date'].map(dt.datetime.toordinal)
maha.head()
```

```python
#Getting dependent and independent variable
x=maha['date']
y=maha['confirmed']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
maha.tail()
```
